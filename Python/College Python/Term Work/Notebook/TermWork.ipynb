```python
#Objective:

#To learn and implement variable defination and declaration
#To learn different datatypes in python

#integer variable
my_int = 5
print(my_int,"is of type",type(my_int),"and id is",id(my_int),"\n")

#float variable
my_float = 5.5
print(my_float,"is of type",type(my_float),"and id is",id(my_float),"\n")

#string variable
my_string = "Hello"
print(my_string,"is of type",type(my_string),"and id is",id(my_string),"\n")

#boolean variable
my_bool = True
print(my_bool,"is of type",type(my_bool),"and id is",id(my_bool),"\n")

#list variable
my_list = [1,2,3]
print(my_list,"is of type",type(my_list),"and id is",id(my_list),"\n")

#tuple variable
my_tuple = (1,2,3)
print(my_tuple,"is of type",type(my_tuple),"and id is",id(my_tuple),"\n")

#set variable
my_set = {1,2,3,"a","b","c"}
print(my_set,"is of type",type(my_set),"and id is",id(my_set),"\n")

#dictionary variable
my_dict = {"a":1, "b":2, "c":3}
print(my_dict,"is of type",type(my_dict),"and id is",id(my_dict),"\n")

#complex variable
my_complex = 1+2j
print(my_complex,"is of type",type(my_complex),"and id is",id(my_complex),"\n")


#algo:

# Define variables with different data types.
# Print each variable's value, data type using type(), and memory address using id() in a single line.
# Repeat step 2 for each variable.
# End program.
```

```python
#Objective: 
# To learn and implements python operators

#Assigning values to variables
a = 10
b = 20

'''Assignment operator'''

print("\nOutput of Assignment operator:\n")
#addition
print("the value of a + b is ", 10+20)
#subtraction
print("the value of a - b is ", 10-20)
#multiplication
print("the value of a * b is ", 10*20)
#division
print("the value of a / b is ", 10/20)
#floor division
print("the value of a // b is", 10//20)
#exponent  
print("the value of a ** b is", 10**20)
#modulus
print("the value of a % b is", 10%20)
#float modulus
print("the value of a % b is", 10.5%20.5)

"""logical operator"""

print("\nOutput of Logical operator:\n")
#and
print("the value of a and b is", 10 and 20)
#or
print("the value of a or b is", 10 or 20)
#not
print("the value of not a is", not 10)

"""Relational operator"""

print("\nOutput of Relational operator:\n")
#greater than
print("the value of a > b is", 10 > 20)
#less than
print("the value of a < b is", 10 < 20)
#greater than or equal to
print("the value of a >= b is", 10 >= 20)
#less than or equal to
print("the value of a <= b is", 10 <= 20)
#equal to
print("the value of a == b is", 10 == 20)
#not equal to
print("the value of a != b is", 10 != 20)

"""Bitwise operator"""

print("\nOutput of Bitwise operator:\n")
#bitwise and
print("the value of a & b is", 10 & 20)
#bitwise or
print("the value of a | b is", 10 | 20)
#bitwise not
print("the value of ~a is", ~10)
#bitwise xor
print("the value of a ^ b is", 10 ^ 20)
#bitwise right shift
print("the value of a >> b is", 10 >> 20)
#bitwise left shift
print("the value of a << b is", 10 << 20)

'''Identity Operators'''

# is Identity
print("\nOutput of Identity operator:\n")
a=10
b=20
print(a is b)

# is not Identity
a=10
b=20
print(a is not b)

'''Membership Operators'''

# in Membership
print("\nOutput of Membership operator:\n")
a=10
b=[10,20,30] #
print(a in b)

# not in Membership
a=10
b=[10,20,30] #list
print(a not in b)

#Algorithm:

# Assign values 10 and 20 to variables a and b respectively.

# Output the results of the assignment operation using the print() function with appropriate messages.

# Use the print() function with appropriate messages to output the results of the logical operators 'and', 'or', and 'not' using the values of a and b.

# Use the print() function with appropriate messages to output the results of the relational operators 'greater than', 'less than', 'greater than or equal to', 'less than or equal to', 'equal to', and 'not equal to' using the values of a and b.

# Use the print() function with appropriate messages to output the results of the bitwise operators 'bitwise and', 'bitwise or', 'bitwise not', 'bitwise xor', 'bitwise right shift', and 'bitwise left shift' using the values of a and b.

# Use the 'is' operator to check the identity of the variables a and b, and output the results using the print() function with appropriate messages.

# Use the 'is not' operator to check the identity of the variables a and b, and output the results using the print() function with appropriate messages.

# Use the 'in' operator to check if the value of variable a is in the list [10,20,30], and output the result using the print() function with appropriate messages.

# Use the 'not in' operator to check if the value of variable a is not in the list [10,20,30], and output the result using the print() function with appropriate messages.
```
